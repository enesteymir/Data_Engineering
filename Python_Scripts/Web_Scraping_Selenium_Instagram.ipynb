```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
# create a browser 
browser = webdriver.Chrome()
```

### Login Instagram

#### Login Instagram by using Path

```python
url = "https://instagram.com"
```

```python
# open the url in the browser
browser.get(url)
time.sleep(5)
```

```python
# Get the element's path for username section
username = browser.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[1]/div/label/input' )
```

```python
# send the username to the instagram
username.send_keys('username')
```

```python
password = browser.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[2]/div/label/input' )
```

```python
# send the username to the instagram
password.send_keys('password')
```

```python
# Click the login button
login_button = browser.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[3]/button/div' )
login_button.click()
```

```python
# Click the followers section
followers_button = browser.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a/div' )
followers_button.click()
```

We need to find the related div that consists all the followers. Then write the proper javascript commands to scroll down.



```python
jscommands = """followers = document.querySelector(".isgrP") ;
                followers.scrollTo(0, followers.scrollHeight) ;
                var lenOfPage = followers.scrollHeight ;
                return lenOfPage ;              
             """
```

```python
# Scroll Down Commands. It will scroll down until reach the bottom.
lenOfPage = browser.execute_script(jscommands)
match=False
while(match==False):
    lastCount = lenOfPage   
    time.sleep(1)
    lenOfPage = browser.execute_script(jscommands)
    if lastCount == lenOfPage:
        match=True
```

### Pull Followers

> The class name is same for all followers when we look to the element names. So we can use class name.

> element class="Jv7Aj mArmR MqpiF" means that it consists of three different classes. So you can not write it as a whole class. 



```python
followers = browser.find_elements(By.CLASS_NAME,  'MqpiF' )
```

```python
follower_list = []
for i in followers:
    follower = i.text
    follower_list.append(follower)
```
